```python
#remove sims with tags as Nans
sim_tags = sim['tags'].dropna()

#Find sims with minSpeed in tags list
min_speed_sims = sim_tags[sim_tags.apply(lambda x: 'baseConfig' in x)].index
min_speed_sims = sim.loc[min_speed_sims]

min_speed_sims
```

```python
#Compute avg xtes
min_speed_sims = min_speed_sims.copy()
min_speed_sims['avg_xtes'] = min_speed_sims['out.xtes'].apply(lambda x: sum(abs(i) for i in x)/len(x) if len(x)>0 else None)

#Compute fps
min_speed_sims['fps'] = min_speed_sims['out.iterations'] / min_speed_sims['out.elapsedTime']

#set a max min speed to 30
min_speed_sims.loc[min_speed_sims['in.minSpeed'] > 30, 'in.minSpeed'] = 30
min_speed_sims[['in.minSpeed', 'avg_xtes']]
```
